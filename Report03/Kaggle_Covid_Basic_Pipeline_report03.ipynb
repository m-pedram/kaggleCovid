# Reading and preprocessing the Data

Cleaning the data has been carried out manually. The already cleaned table has been uploaded in a Github Repository.

```python
# read data from Google Drive
# from google.colab import drive 
# drive.mount('/content/drive')

data = pd.read_csv('https://raw.githubusercontent.com/m-pedram/kaggleCovid/main/Report03/rawData.csv')
```

To fill the missing values several approaches have been implemented:

"padding": this is a simple methode sugested by the dataset ceators. The next or the previous entries have been copied. In this regard, the Forward filling (Padding) has been favored over backfilling.

"most_frequent" "mean" "median" : Other univariate feature imputation methdes provided by sklearn.

"multivariate": An IterativeImputer class which is a more sophisticated approach. However, this estimator is still experimental in sklearn library.

"KNN": provides imputation for filling in missing values using the k-Nearest Neighbors approach.

```python
imputaion_methode = "KNN" # "padding" "most_frequent" "mean" "median" "multivariate" "KNN"
table = np.array([0])
```

```python
if imputaion_methode == "padding":
  filledData = data.query("PATIENT_VISIT_IDENTIFIER == 0").fillna(method='pad')
  filledData = filledData.query("PATIENT_VISIT_IDENTIFIER == 0").fillna(method='backfill')
  for patient_identifier in range(1,255):
    tempDf = data.query("PATIENT_VISIT_IDENTIFIER == @patient_identifier").fillna(method='pad')
    tempDf2 = tempDf.query("PATIENT_VISIT_IDENTIFIER == @patient_identifier").fillna(method='backfill')
    filledData = pd.concat([filledData, tempDf2])
  table = filledData.to_numpy()
```

```python
if imputaion_methode in ["most_frequent", "mean", "median"]:
  imp = SimpleImputer(missing_values=np.nan, strategy= imputaion_methode)
  imp.fit(data.to_numpy()[0:4,])
  table = imp.transform(data.to_numpy()[0:4,])

  for patient in range(1, int((data.shape[0])/4)):
    imp.fit(data.to_numpy()[patient*4:patient*4+4,])
    table = np.vstack((table, imp.transform(data.to_numpy()[patient*4:patient*4+4,])))
```

```python
if imputaion_methode == "multivariate":
  imp = IterativeImputer(max_iter=100, random_state=0)
  imp.fit(data.to_numpy()[0:4,])
  table = imp.transform(data.to_numpy()[0:4,])

  for patient in range(1, int((data.shape[0])/4)):
    imp.fit(data.to_numpy()[patient*4:patient*4+4,])
    table = np.vstack((table, imp.transform(data.to_numpy()[patient*4:patient*4+4,])))  
```

```python
if imputaion_methode == "KNN":
  imputer = KNNImputer(n_neighbors=2, weights="uniform")
  table = imputer.fit_transform(data.to_numpy()[0:4,])

  for patient in range(1, int((data.shape[0])/4)):
    table = np.vstack((table, imputer.fit_transform(data.to_numpy()[patient*4:patient*4+4,])))  
```

```python
if(np.count_nonzero(np.isnan(table)) == 0 ): # just to make sure there is no nan values.
   clean_table = table
```

```python
#functions

def random_imputation(df, feature):
    number_missing = df[feature].isnull().sum()
    observed_values = df.loc[df[feature].notnull(), feature]
    df.loc[df[feature].isnull(), feature + '_imp'] = np.random.choice(observed_values, number_missing, replace=True)
    return df

def knn_imputation(X):
  imputer = KNNImputer(n_neighbors=2, weights="uniform")
  return imputer.fit_transform(X)


def logistic_regression(X_tr, X_ts, y_tr, cWeight):
    clf = LogisticRegression(random_state=0, max_iter= 1000, class_weight=cWeight).fit(X_tr, y_tr.reshape(-1, ))
    return clf.predict(X_ts)


def support_vector_machine(X_tr, X_ts, y_tr,cWeight):
    # default kernel is 'RBF'
    clf = make_pipeline(StandardScaler(), SVC(gamma='auto', kernel='rbf', class_weight=cWeight))
    clf.fit(X_tr, y_tr.reshape(-1, ))
    return clf.predict(X_ts)


def metrics_report(y_true, y_pred):
  accuracy = accuracy_score(y_test, y_pred)
  f1 = f1_score(y_test, y_pred)
  balancedAccuracy = balanced_accuracy_score(y_test, y_pred)
  precision = average_precision_score(y_test, y_pred)
  return accuracy, f1, balancedAccuracy, precision
```
